```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
import numpy as np
```

```python
movies = pd.read_csv("final_movies.csv")
#movies
```

```python
links = pd.read_csv("final_links.csv")
links
```

```python
ratings = pd.read_csv("final_ratings.csv")
#ratings
```

```python
links_array = np.array(links)
```

```python
img_urls = []
for i in range(len(links_array)):
    film_id = str(links_array[i][1])[:-2]
    #print(film_id)
    try:
        url = ''
        if len(film_id) == 4:
            url = 'http://www.imdb.com/title/tt000' + film_id
        elif len(film_id) == 5:
            url = 'http://www.imdb.com/title/tt00' + film_id
        elif len(film_id) == 6:
            url = 'http://www.imdb.com/title/tt0' + film_id
        else :
            url = 'http://www.imdb.com/title/tt' + film_id
        soup = BeautifulSoup(urlopen(url).read())
        link = soup.find(itemprop="image")
        src_path = link["src"]
        print(src_path)
        img_urls.append(src_path)
    except:
        print(movies['title'][i] , film_id , "Not Found---------------------------------------------------------")
        img_urls.append("None" + film_id)
```
